```python
# %%
import pandas as pd
import re

# 📦 Reunimos todos los DataFrames cuyo nombre empieza por df_melted_*
dfs_named = [(name, df) for name, df in globals().items()
             if name.startswith("df_melted_") and isinstance(df, pd.DataFrame)]

# 📅 Orden de los meses en español
MONTHS = {
    "enero": 1, "febrero": 2, "marzo": 3, "abril": 4, "mayo": 5, "junio": 6,
    "julio": 7, "agosto": 8, "septiembre": 9, "octubre": 10,
    "noviembre": 11, "diciembre": 12
}

# 🔧 Función para ordenar por nombre del DataFrame: extrae mes, año y isla
def sort_key(name_df):
    name, _ = name_df
    mes, año, isla = None, None, "unknown"

    # Ejemplo de nombre: df_melted_fuerteventura_enero_2024
    m = re.search(r'_(enero|febrero|marzo|abril|mayo|junio|julio|agosto|septiembre|octubre|noviembre|diciembre)_(\d{4})', name)
    i = re.search(r'df_melted_([a-z]+)_', name)
    if m:
        mes, año = m.groups()
    if i:
        isla = i.group(1).capitalize()

    mes_num = MONTHS.get(mes, 99)
    año = int(año) if año else 9999
    return (año, mes_num, isla)

# 📊 Ordenamos por año, mes e isla
dfs_named_sorted = sorted(dfs_named, key=sort_key)

print("📅 Orden de unión:\n")
for i, (n, _) in enumerate(dfs_named_sorted, 1):
    print(f"{i:02d}. {n}")

# 🧩 Unimos todos los DataFrames en una sola tabla
df_all = pd.concat([df for _, df in dfs_named_sorted], ignore_index=True)

# 📆 Separación de 'Mes_Año' en mes, año y fecha
df_all["Mes"] = df_all["Mes_Año"].str.extract(r"^([a-zñA-Z]+)")[0].str.lower()
df_all["Año"] = df_all["Mes_Año"].str.extract(r"(\d{4})").astype(int)
df_all["MesNum"] = df_all["Mes"].map(MONTHS)
df_all["Fecha"] = pd.to_datetime(dict(year=df_all["Año"], month=df_all["MesNum"], day=1))

# 📍 Ordenación final: isla → año → mes
df_all_sorted = df_all.sort_values(by=["Isla", "Año", "MesNum"])

# 📁 Guardamos el conjunto completo
df_all_sorted.to_csv("result.csv", index=False)

# 🎯 Filtramos únicamente TOTAL PASAJEROS
col_nazwa = df_all_sorted.columns[0]  # primera columna (con los nombres)
mask = df_all_sorted[col_nazwa].astype(str).str.upper() == "TOTAL PASAJEROS"
df_total = df_all_sorted.loc[mask, ["Isla", "Fecha", "Mes", "Año", "MesNum", "Pasajeros"]]

df_total.to_csv("result_total.csv", index=False)

print(f"\n✅ Se han unido {len(dfs_named_sorted)} DataFrames.")
print(f"➡️ Resultado: {df_all_sorted.shape[0]} filas, {df_all_sorted.shape[1]} columnas.")
print(f"💾 Guardado 'result.csv' y 'result_total.csv'")
```
